```python
import pandas as pd
from os import walk
```

```python
def get_grammar(g_directory='../../data/grammar'):
    file_names = next(walk(g_directory), (None, None, []))[2]

    grammar = []
    for file in file_names[1:]:
        grammar.append(pd.read_csv(g_directory+'/'+file, header=None))

    return grammar
```
